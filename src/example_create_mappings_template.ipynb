```python

import rdflib
from rdflib import Graph, Namespace, URIRef, Literal, BNode
from rdflib.namespace import NamespaceManager
from rdflib.namespace import RDF, RDFS, SKOS, XSD, OWL

import pandas as pd
```

# Load the CD description

```python
source_file_1 = '../data/output/cdi_profile_EMP.ttl'
g1 = Graph()
g1.parse(source_file_1, format="turtle")

g2 = Graph()
source_file_2 = '../data/output/cdi_profile_GDP.ttl'
g2.parse(source_file_2, format="turtle")
```

## Query the RDF graph for nodes with enumeration codes: 


```python
query = """
    SELECT ?iri ?label ?scheme
    WHERE {
        ?iri a skos:Concept .
        ?iri rdfs:label ?label .
        ?iri skos:inScheme ?scheme .
    }
"""
```

```python
enumerations_1 = [(str(row.iri), str(row.label), str(row.scheme)) for row in g1.query(query)]
enumerations_1 = pd.DataFrame(enumerations_1, columns=["IRI", "Label", "Enumeration"])
enumerations_1
```

```python
enumerations_2 = [(str(row.iri), str(row.label), str(row.scheme)) for row in g2.query(query)]
enumerations_2 = pd.DataFrame(enumerations_2, columns=["IRI", "Label", "Enumeration"])
enumerations_2
```

```python
# Save the two DataFrames as separate sheets in a single Excel file
with pd.ExcelWriter("../data/output/mapings_EMP_GDP.xlsx", engine='xlsxwriter') as writer:
    enumerations_2.to_excel(writer, sheet_name='EMP', index=False)
    enumerations_2.to_excel(writer, sheet_name='GDP', index=False)
```
